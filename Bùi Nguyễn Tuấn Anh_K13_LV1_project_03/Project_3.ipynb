```python
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/yinghaoz1/tmdb-movie-dataset-analysis/master/tmdb-movies.csv"
df = pd.read_csv(url)
```

```python
df.info()
```

```python
#1: Sắp xếp các bộ phim theo ngày phát hành giảm dần rồi lưu ra một file mới
sorted_df = df.sort_values('release_date')
sorted_df.head()
```

```python
#2: Lọc ra các bộ phim có đánh giá trung bình trên 7.5 rồi lưu ra một file mới
filtered_df = df[df['vote_average'] >= 7.5]
filtered_df.head(10)
```

```python
#3: Tìm ra phim nào có doanh thu cao nhất và doanh thu thấp nhất
lowest_revenue = df.sort_values('revenue', ascending=True).head(1)
highest_revenue = df.sort_values('revenue', ascending=False).head(1)
print(lowest_revenue, highest_revenue)
```

```python
#4: Tính tổng doanh thu tất cả các bộ phim
total_revenue = df['revenue'].sum()
print(total_revenue)
```

```python
#5: Top 10 bộ phim đem về lợi nhuận cao nhất
top_10_movies = df.sort_values('revenue', ascending=False).head(10)
print(top_10_movies)
```

```python
#6: Đạo diễn nào có nhiều bộ phim nhất và diễn viên nào đóng nhiều phim nhất
director_movies = df.groupby('director')['id'].count().sort_values(ascending=False).head(1)
print(director_movies)
actor = df.groupby('cast')['id'].count().sort_values(ascending=False).head(1)
print(actor)
```

```python
#7: Thống kê số lượng phim theo các thể loại. Ví dụ có bao nhiêu phim thuộc thể loại Action, bao nhiêu thuộc thể loại Family, ….
genres = df['genres'].str.split('|')
genres_exploded = genres.explode('genres').value_counts()
print(genres_exploded)
```
